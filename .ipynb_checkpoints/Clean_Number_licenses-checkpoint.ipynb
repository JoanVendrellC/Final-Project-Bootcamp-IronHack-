Checking the response status for the url_2021

```python
response = requests.get(url_2021)
response.status_code
```

Printing the content of the url.

```python
soup = BeautifulSoup(response.text, "html.parser")
soup
```

Printing just the table.

```python
License_table=soup.find('table',class_ = "ApartNum xs Cols5")
License_table
```

```python
List_activities = []

for activity in License_table.find_all('tbody'):
    rows = activity.find_all('tr')
    for row in rows:
        List_activities.append(row.find('th').text)
        
        
List_licences=[]

for activity in License_table.find_all('tbody'):
    rows = activity.find_all('tr')
    for row in rows:
        List_licences.append(row.find_all('td')[3].text)

```

Creating a dataframe

```python
Licenses_2021 = pd.DataFrame({"Sport":List_activities,
                       "Total_licences":List_licences,
                      })
```

Getting just the value for Ciclism and creating a new column for the year.

```python
Licenses_2021=Licenses_2021[Licenses_2021['Sport']=='Ciclisme']
Licenses_2021['Year'] = ['2021']
Licenses_2021.head()
```
